```python
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as mpl
import numpy as np

data = pd.read_csv('../data/house_pricing.csv')
data.info()
sb.relplot(x="LotArea", y="SalePrice", data=data)
data[:]
```

```python
dataframe = pd.DataFrame(data, columns = ['SalePrice', 'LotArea'])
affinage = dataframe[(dataframe['LotArea'] < 20000) & (dataframe['SalePrice'] < 500000)]
sb.relplot(x="LotArea", y="SalePrice", data = affinage)
```
